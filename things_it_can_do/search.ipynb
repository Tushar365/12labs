```python
import os
from dotenv import load_dotenv
from twelvelabs import TwelveLabs

# 1. Load API Key
load_dotenv()
apiKey = os.getenv("TL_API_KEY")

client = TwelveLabs(api_key=apiKey)
index_id = "69538a69db0246c06ce302ac" 

print(f"Using Index ID: {index_id}")
```

```python
search_pager = client.search.query(
    index_id="69538a69db0246c06ce302ac",
    query_text="riding a mountain bike",
    search_options=["visual", "audio"],
    
    # operator="or"
)

print("Search results:")
for clip in search_pager:
    print(
        f" video_id {clip.video_id} score={clip.score} start={clip.start} end={clip.end} confidence={clip.confidence}"
    )
```

```python
search_pager = client.search.query(
    index_id="69538a69db0246c06ce302ac",
    query_text="riot",
    search_options=["visual", "audio"],
    operator="or",
    group_by="video",
    sort_option="score"
)

print("Search results:")
for group in search_pager:
    if group.clips is None:
        continue
    print(f"  Video ID: {group.id}")
    for clip in group.clips:
        print(
            f"  rank={clip.rank} start={clip.start} end={clip.end}"
        )
```
